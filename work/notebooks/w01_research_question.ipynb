## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

Loading the starter dataset to check how big this problem actually is — how many pages are
declining, and whether a simple rule captures them well.

The Output:

* Total pages in starter dataset: 30000

* % of pages currently declining: 54.2%

* % of pages that are stale AND still getting real traffic: 0.1%


```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

import pandas as pd

df = pd.read_csv("/content/content_refresh_anonymized.csv")

pct_declining = (df["trend_direction"].str.lower() == "down").mean()
pct_stale_visible = ((df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)).mean()

print(f"Total pages in starter dataset: {df.shape[0]}")
print(f"% of pages currently declining: {pct_declining:.1%}")
print(f"% of pages that are stale AND still getting real traffic: {pct_stale_visible:.1%}")
```
